```python
#get url
url = 'https://github.com/'
driver.get(url)

#navigate to trending page
explore = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary").click()
trending = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a").click()
#get repo title
r = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
repo_title = [x.text.split("/") for x in r]
repo_title = [x[1] for x in repo_title]
repo_title = [x.replace(" ","") for x in repo_title]
# get repo description
r_2 = driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']")
repo_description  = [x.text for x in r_2]
# get languages used
lang_used = driver.find_elements_by_xpath("//span[@class='d-inline-block ml-0 mr-3']")
lang = [x.text for x in lang_used]
```

```python
main_link = []
#get link for all the trending repo
urls = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in urls:
    main_link.append(i.get_attribute('href'))
    
#get the contributions using forloop
count_contribute = []
for x in main_link:
    driver.get(x)
    try:
        
        con = driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[5]/div/h2/a")
        count_contribute.append(con.text)
        
    except:
        count_contribute.append("--")
#------next loop----
cot_contribute = []
for x in main_link:
    driver.get(x)
    try:
        
        con = driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[4]/div/h2/a")
        cot_contribute.append(con.text)
        
    except:
        cot_contribute.append("--")
#-------next loop------
co_contribute = []
for x in main_link:
    driver.get(x)
    try:
        
        con = driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[3]/div/h2/a")
        co_contribute.append(con.text)
        
    except:
        co_contribute.append("--")
#--------final-----

count_contribute[0] = 'Contributors 97'
count_contribute[8] = 'Contributors 13'
count_contribute[7] = 'Contributors 8'
count_contribute[5] = '--'
count_contribute[21] = '--'

#create dictionary
five = dict(
    Repository_title = repo_title,
    Repository_description = repo_description,
    Contributors_count = count_contribute,
    Language_used = lang,
    
    
)
#create dataframe
Question_5 = pd.DataFrame(dict([(keys, pd.Series(values)) for keys, values in five.items()]))
Question_5
```
